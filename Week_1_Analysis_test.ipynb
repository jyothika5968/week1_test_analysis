```python
##Task-1
# Swizzy and Zomato Food delivery analysis

# a food delivery comapny wants to improve efficiency and customer satisfaction

# data
# order time,delivery time,delivery partner,customer ratings


# Identify
#avg delivery time
#peak order hours
# most delayed areas
# top rated delivery partners

#Create visualization
#charts,performance report
```

```python
# Data Set
import statistics
order_time = [3.30,4.40,6.15,7.20,5.50]
delivery_time = [2,3,4,5,6]
delivery_partner = ["a","b","c","d","e"]
customer_ratings = [4,5,3,2,4]
print("order_time",order_time)
print("delivery_time",delivery_time)
print("delivery_partner",delivery_partner)
print("customer_ratings",customer_ratings)
```

```python
#avg delivery time
avg_delivery_time = sum(delivery_time)/len(delivery_time)
print("avg_delivery_time",avg_delivery_time)
```

```python
#peak order hours
peak_order_hours = max(order_time)
print("peak_order_hours",peak_order_hours)
```

```python
# most delayed areas
max_delivery_time_index = delivery_time.index(max(delivery_time))
most_delayed_partner = delivery_partner[max_delivery_time_index]
print("most_delayed_partner", most_delayed_partner)
```

```python
#top rated delivery partners
max_rating_index = customer_ratings.index(max(customer_ratings))
top_rated_partner = delivery_partner[max_rating_index]
print("top_rated_partner", top_rated_partner)
```

```python
# Visualization
# create Bar Diagram
import matplotlib.pyplot as plt
plt.bar(delivery_partner,customer_ratings)
plt.xlabel("delivery_partner")
plt.ylabel("customer_ratings")
plt.title("Top rated delivery partners")
plt.show()
```

```python
# Performance Report
print("Performance Report")
print("Average Delivery Time:", avg_delivery_time)
print("Peak Order Hour:", peak_order_hours)
print("Most Delayed Partner:", most_delayed_partner)
print("Top Rated Partner:", top_rated_partner)
```

```python
# Customer Satisfaction For Delivery Time and Customer ratings
plt.bar(delivery_partner,delivery_time)
plt.xlabel("delivery_partner")
plt.ylabel("delivery_time")
plt.title("Customer Satisfaction For Delivery Time")
plt.show()
```

```python
# create pie Diagram
plt.pie(customer_ratings,labels=delivery_partner)
plt.title("Customer Ratings")
plt.show()
```

```python
##Task-2
# customer Behaviour

# Analyze
# Product purchases
#card abandment
# Customer Rating
#payment methods

# Identify
# Most sold products
# High Range Customers
# most Used Payment Method
# Products with poor Ratings

 # Data Set
product_purchase = [10,12,14,16,10]
card_abandment = [25,33,65,72,60]
Customer_Rating = [18,37,54,86,93]
payment_methods = ["Credit Card","Debit Card","UPI","Net Banking","UPI"]
print("product_purchase",product_purchase)
print("card_abandment",card_abandment)
print("Customer_Rating",Customer_Rating)
print("payment_methods",payment_methods)
```

```python
##Task-3

#bank fraud detectio analysis

# Data Set
#transaction amount
#location
#time
#account activity


#identify
#unusual transactionas
#high risk accounts
#fraud pattern
```

```python
# Data Set
import statistics
import matplotlib.pyplot as plt
Trasanction_amount = [2000,3500,4500,2700]
location = ['Banglore','Madanapalle','Tamilnadu','chennai']
time = [5,7,9,12]
account_activity = [10,15,20,30]
print("Trasanction_amount :",Trasanction_amount)
print("location:",location)
print("time:",time)
print("account_activity:",account_activity)
```

```python
#unusual transactions
max_amount = max(Trasanction_amount)
print("max_amount",max_amount)
```

```python
#high risk accounts
for i in range(len(account_activity)):
    if account_activity[i] > 16:
        print(location[i])
```

```python
#fraud pattern
for i in range(len(time)):
    if time[i] < 5:   # midnight fraud (0–4 hours)
        print("Fraud at odd time:", location[i])
```

```python
# Visualisation
plt.bar(location,Trasanction_amount)
plt.xlabel("location")
plt.ylabel("Trasanction_amount")
plt.title("Suspicious Transactions")
plt.show()
```

```python
plt.scatter(Trasanction_amount,time)
plt.xlabel("Trasanction_amount")
plt.ylabel("time")
plt.title("Transactions done based on the time")
plt.show()
```

```python
## Task 4
#   Supermarket inventory & sales and analysis
# a super market wants to reduce product wastage and improve sock management


#Data Set
#daily sales
# product stock
#expiry date
# seasonal demand

# Identify
# fast moving products
# slow moving product
# out of stock trends


# Suggest
```

```python
# Data set
daily_sales = [100,200,300,400,500]
product_stock = [50,30,40,20,10]
expiry_date = [2,15,25,7,21]
seasonal_demand = [10,15,20,25,30]
print("daily_sales:",daily_sales)
print("product_stock:",product_stock)
print("expiry_date:",expiry_date)
print("seasonal_demand:",seasonal_demand)
```

```python
#fast moving products
max_sales = max(daily_sales)
print("max_sales",max_sales)
for i in range(len(daily_sales)):
    if daily_sales[i] == max_sales:
        print("Fast moving product:",i+1)
```

```python
# slow moving product
min_sales = min(daily_sales)
print("min_sales",min_sales)
for i in range(len(daily_sales)):
    if daily_sales[i] == min_sales:
        print("Slow moving product:",i+1)
```

```python
# Corrected multi-line code
for i in range(len(product_stock)):
    if product_stock[i] == 10:
        print("Out of stock product expiry date:", expiry_date[i])
```

```python
# Visualization
import matplotlib.pyplot as plt
plt.plot(daily_sales,product_stock)
plt.xlabel("daily_sales")
plt.ylabel("product_stock")
plt.title("Fast moving products")
plt.show()
```
